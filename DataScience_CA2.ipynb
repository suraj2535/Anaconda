```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
# Check for missing values
df.isnull().sum() 
```

```python
# Filter out products with zero ratings if necessary
df_cleaned = df[df['rating'] != 0]
df_cleaned
```

```python
# Summary statistics for numeric columns
df_cleaned.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot rating distribution
sns.histplot(df_cleaned['rating'], bins=5)
plt.title('Distribution of Product Ratings')
plt.show()
```

```python
sns.histplot(df_cleaned['discount_percent'], bins=10)
plt.title('Distribution of Discount Percentages')
plt.show()
```

```python
avg_discount_by_brand = df_cleaned.groupby('brand_name')['discount_percent'].mean().reset_index()
avg_discount_by_brand = avg_discount_by_brand.sort_values(by='discount_percent', ascending=False)
avg_discount_by_brand.head()
```

```python
total_ratings_by_brand = df_cleaned.groupby('brand_name')['rating_count'].sum().reset_index()
total_ratings_by_brand = total_ratings_by_brand.sort_values(by='rating_count', ascending=False)
total_ratings_by_brand.head()
```

# Questions and Analysis

# 1. Which product has the highest rating?

```python
highest_rated_product = df.loc[df['rating'].idxmax()]
highest_rated_product[['product_name', 'rating']]
```

# 2. What is the total number of products available in each category (product_tag)?

```python
products_per_category = df['product_tag'].value_counts()
products_per_category
```

# 3. Which brand offers products with the highest average discount?

```python
highest_avg_discount_brand = df.groupby('brand_name')['discount_percent'].mean().idxmax()
highest_avg_discount_brand
```

# 4. Calculate the total revenue lost due to discounts for all products.

```python
df['revenue_lost'] = df['marked_price'] - df['discounted_price']
total_revenue_lost = df['revenue_lost'].sum()
total_revenue_lost
```

# 5. What is the median discounted price across all products?

```python
median_discounted_price = np.median(df['discounted_price'])
median_discounted_price
```

# 6. Generate an array of the top 5 discounted products.

```python
top_5_discounts = np.array(df.nlargest(5, 'discount_percent')[['product_name', 'discount_percent']])
top_5_discounts
```

# 7. What is the average percentage discount for products priced above ₹1000?

```python
high_price_discounts = df[df['marked_price'] > 1000]['discount_percent'].mean()
high_price_discounts
```

# 8. Which brand has the highest variance in product prices?

```python
price_variance_per_brand = df.groupby('brand_name')['marked_price'].var().idxmax()
price_variance_per_brand
```

# 9. Visualize the distribution of marked_price using Seaborn.

```python
sns.histplot(df['marked_price'], bins=10)
plt.title('Distribution of Marked Prices')
plt.show()
```

# 10. Create a box plot to show the distribution of discount_percent across different product categories (product_tag).

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='product_tag', y='discount_percent', data=df)
plt.xticks(rotation=90)
plt.title('Discount Percent Across Product Categories')
plt.show()
```

# 11. Create a scatter plot to show the relationship between marked_price and discounted_price.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='marked_price', y='discounted_price', data=df)
plt.title('Marked Price vs Discounted Price')
plt.show()
```

# 12. Plot a bar chart to show the number of products available per brand.

```python
products_per_brand = df['brand_name'].value_counts()

plt.figure(figsize=(10, 6))
products_per_brand.plot(kind='bar', color='skyblue')
plt.title('Number of Products Per Brand')
plt.ylabel('Number of Products')
plt.show()
```

# 13. Create a pie chart to show the percentage of products with and without ratings (rating_count > 0).

```python
has_ratings = df['rating_count'] > 0
ratings_count = has_ratings.value_counts()

plt.figure(figsize=(6, 6))
plt.pie(ratings_count, labels=['With Ratings', 'Without Ratings'], autopct='%1.1f%%', colors=['lightgreen', 'orange'])
plt.title('Percentage of Products with and without Ratings')
plt.show()
```

# 14. Create a line plot to show the change in discount_percent for the top 5 products with the highest ratings.

```python
top_5_high_rated = df.nlargest(5, 'rating')[['product_name', 'discount_percent']]

plt.figure(figsize=(8, 5))
plt.plot(top_5_high_rated['product_name'], top_5_high_rated['discount_percent'], marker='o', color='blue')
plt.title('Discount Percent for Top 5 High-Rated Products')
plt.ylabel('Discount Percent')
plt.xticks(rotation=45)
plt.show()
```

# 15. Display a heatmap of correlations between numerical features (rating, marked_price, discounted_price, etc.).

```python
corr = df[['rating', 'rating_count', 'marked_price', 'discounted_price', 'discount_percent']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
