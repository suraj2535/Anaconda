```python
df['new_Date'] = pd.to_datetime(df['Order Date'], format='%d%m%Y', errors='coerce')
```

```python
df['Day_of_Week'] = df['new_Date'].dt.day_of_week
df['Day_of_Week'].head()
```

```python
def WeekNum(day):
    if day < 1 or day > 31:
        raise ValueError("Day should be between 1 and 31")
    if day <= 7:
        return 'week1'
    elif day <= 14:
        return 'week2'
    elif day <= 21:
        return 'week3'
    elif day <= 28:
        return 'week4'
    else:
        return 'week5'
```
